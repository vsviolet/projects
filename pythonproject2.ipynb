## Look at the times and days with most accidents 

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
df.Start_Time
```

```python
# Time of day where accidents are most common 

sns.histplot(df.Start_Time.dt.hour, bins=24)
```

#### Highest percentage of accidents happen between 4 and 6 pm. Likely because people are leaving work and rushing to get home 

```python
# Days of the week with most accidents

plt.figure(figsize=(10,5))
sns.histplot(df.Start_Time.dt.day_name())
plt.xlabel('Days of the week')
plt.title('Accidents During the Week')
```

#### More accidents during the weekdays than the weekends

```python
# Check out when accidents occur on Sunday 

sundays_accidents = df.Start_Time[df.Start_Time.dt.day_name() == 'Sunday']
sns.histplot(sundays_accidents.dt.hour, bins=24)
```

#### Most Sunday accidents occur from noon to 8 pm.

```python
# Monday accidents 

mondays_accidents = df.Start_Time[df.Start_Time.dt.day_name() == 'Monday']
sns.histplot(mondays_accidents.dt.hour, bins=24)
```

#### Most Monday accidents are in the morning around 7 am and from 3 to 5 pm

```python
months_accidents = df.Start_Time.dt.month_name().value_counts()
months_accidents
```

```python
plt.figure(figsize=(14,5))
sns.barplot(y=months_accidents, x=months_accidents.index)
plt.ylabel('Number of Accidents')
plt.title('Accidents Each Month')
```

#### Most accidents occur in December, November and October 

```python
# See monthly data for 2019 only 

df_2019 = df[df.Start_Time.dt.year == 2019]
sns.histplot(df_2019.Start_Time.dt.month, bins=12)
```

```python
yearly_accidents = df.Start_Time.dt.year.value_counts()
yearly_accidents
```

```python
sns.lineplot(data = yearly_accidents)
plt.ylabel('Number of Accidents (in millions)')
plt.xlabel('Year')
```

#### Accidents have been increasing since 2016

```python
# Weather conditions 

weather = df['Weather_Condition'].value_counts()
weather[:15]
```

```python
plt.figure(figsize=(20,5))
sns.barplot(y=weather[:15], x=weather[:15].index)
plt.ylabel('Number of Accidents')
plt.title('Weather condition vs Number of Accidents')
```

#### Strangly enough, most accidents happen when the weather is fair, meaning weather consition isn't as much of a factor as we might think
